This notebook is to show the visualization technique we performed to visualize noise removal from a random sample of three audio files, to see if our denoising technique worked well or not.

```python
import os
import matplotlib.pyplot as plt
import librosa
import numpy as np
```

```python
file2='/kaggle/working/denoised_audio_files/noise_reduced_2090166.wav'
file1='/kaggle/working/audio_files/2090166.wav'
dir2='/kaggle/working/denoised_audio_files'
dir1='/kaggle/working/audio_files'


noisy_file_path = os.path.join(dir1, file1)
y_noisy, sr_noisy = librosa.load(noisy_file_path)


denoised_file_path = os.path.join(dir2, file2)
y_denoised, sr_denoised = librosa.load(denoised_file_path)

# Compute the Short-Time Fourier Transform (STFT) for both audio files
D_noisy = librosa.stft(y_noisy)
D_denoised = librosa.stft(y_denoised)

# Compute the power spectrogram for both audio files
power_noisy = np.abs(D_noisy)**2
power_denoised = np.abs(D_denoised)**2

# Compute the frequencies corresponding to the STFT
freqs = librosa.fft_frequencies(sr=sr_noisy)

# Plot waveform, frequency spectrum, and power spectrum
plt.figure(figsize=(16, 10))

# Plot waveform for noisy audio
plt.subplot(3, 2, 1)
librosa.display.waveshow(y_noisy, sr=sr_noisy, color='#008080')
plt.title('Noisy Audio Waveform')
plt.xlabel('Time/Seconds')
plt.ylabel('Amplitude')

# Plot frequency spectrum for noisy audio
plt.subplot(3, 2, 2)
plt.plot(freqs, np.mean(power_noisy, axis=1), color='blue')
plt.title('Noisy Audio Power Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

# Plot waveform for denoised audio
plt.subplot(3, 2, 3)
librosa.display.waveshow(y_denoised, sr=sr_denoised, color='#008080')
plt.title('Denoised Audio Waveform')
plt.xlabel('Time/Seconds')
plt.ylabel('Amplitude')

# Plot frequency spectrum for denoised audio
plt.subplot(3, 2, 4)
plt.plot(freqs, np.mean(power_denoised, axis=1), color='blue')
plt.title('Denoised Audio Power Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

plt.tight_layout()
plt.show()
```
