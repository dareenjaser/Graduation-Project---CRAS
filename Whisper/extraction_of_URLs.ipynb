## Extracting audio files into 5 directories based on Ratings

```python
data1['Ratings'].unique()
```

```python
# Check length of each audio file of each rating:

ratings = data1['Ratings']

verybad = data1.loc[data1['Ratings'] == 1, 'File'].values #very bad score ratings
bad = data1.loc[data1['Ratings'] == 2, 'File'].values #bad score ratings
med = data1.loc[data1['Ratings'] == 3, 'File'].values #medium score ratings
good = data1.loc[data1['Ratings'] == 4, 'File'].values #good score ratings
full = data1.loc[data1['Ratings'] == 5, 'File'].values #full score ratings

len(verybad), len(bad), len(med), len(good), len(full)
```

```python
# Function to download audio files

def download_audio(link, output_folder):
    try:
        response = requests.get(link)
        if response.status_code == 200:


            file_name = link.split('/')[-1]
            output_path = os.path.join(output_folder, file_name)

            # Save the audio file
            with open(output_path, 'wb') as file:
                file.write(response.content)
            return output_path
        else:
            print(f"Failed to download audio from {link}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error downloading audio from {link}: {e}")
        return None
```

```python
# Create  folder to save the downloaded audio files

vbad_ratings_folder = 'ASR_Data2/V_Bad_Ratings2'

os.makedirs(vbad_ratings_folder, exist_ok=True)

# Download audio files and store their local paths

downloaded_audio_paths = []
for index, audio_link in enumerate(verybad):
    audio_file_path = download_audio(audio_link, vbad_ratings_folder)
    if audio_file_path:
        print(f"Downloaded audio {index + 1}: {audio_file_path}")
        downloaded_audio_paths.append(audio_file_path)

len(downloaded_audio_paths)
```
